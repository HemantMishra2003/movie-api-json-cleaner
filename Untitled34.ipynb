```python
import numpy as np
from PIL import Image
import os
import cv2

PATCH = 64

A_DIR = "/content/EarthData/GoogleEartth/Train/T1"
B_DIR = "/content/EarthData/GoogleEartth/Train/T1"
L_DIR = "/content/EarthData/GoogleEartth/Train/Label"

def extract_features(img1, img2):
    img1 = img1.astype(np.float32)
    img2 = img2.astype(np.float32)

    # --- BASIC FEATURES ---
    diff = img2 - img1                       # 3 features
    ratio = img2 / (img1 + 1e-5)             # 3 features

    # Convert to grayscale
    gray1 = img1.mean(axis=2)
    gray2 = img2.mean(axis=2)

    # Sobel edges → change in structure
    sobel1 = cv2.Sobel(gray1, cv2.CV_64F, 1, 1)
    sobel2 = cv2.Sobel(gray2, cv2.CV_64F, 1, 1)
    sobel_diff = (sobel2 - sobel1)[..., None]    # 1 feature

    # --- NDVI (vegetation) ---
    ndvi1 = (img1[:,:,1] - img1[:,:,0]) / (img1[:,:,1] + img1[:,:,0] + 1e-5)
    ndvi2 = (img2[:,:,1] - img2[:,:,0]) / (img2[:,:,1] + img2[:,:,0] + 1e-5)
    ndvi_diff = (ndvi2 - ndvi1)[..., None]       # 1 feature

    # --- NDWI (water) ---
    ndwi1 = (img1[:,:,1] - img1[:,:,2]) / (img1[:,:,1] + img1[:,:,2] + 1e-5)
    ndwi2 = (img2[:,:,1] - img2[:,:,2]) / (img2[:,:,1] + img2[:,:,2] + 1e-5)
    ndwi_diff = (ndwi2 - ndwi1)[..., None]       # 1 feature

    # --- NDBI (urban/buildings) ---
    ndbi1 = (img1[:,:,0] - img1[:,:,1]) / (img1[:,:,0] + img1[:,:,1] + 1e-5)
    ndbi2 = (img2[:,:,0] - img2[:,:,1]) / (img2[:,:,0] + img2[:,:,1] + 1e-5)
    ndbi_diff = (ndbi2 - ndbi1)[..., None]       # 1 feature

    # FINAL CONCAT
    feat = np.concatenate([
        diff, ratio,
        sobel_diff,
        ndvi_diff,
        ndwi_diff,
        ndbi_diff
    ], axis=-1)

    H, W, C = feat.shape
    return feat.reshape(-1, C)
X_all = []
Y_all = []

files = sorted(os.listdir(A_DIR))

print("Total files:", len(files))

for f in files:
    A = np.array(Image.open(os.path.join(A_DIR, f)))
    B = np.array(Image.open(os.path.join(B_DIR, f)))
    M = np.array(Image.open(os.path.join(L_DIR, f)))

    # Convert mask 0/255 → 0/1
    M = (M > 128).astype(np.uint8)

    feat = extract_features(A, B)
    lab  = M.reshape(-1)

    X_all.append(feat)
    Y_all.append(lab)

X_all = np.concatenate(X_all)
Y_all = np.concatenate(Y_all)

print("Final Feature Shape:", X_all.shape)
print("Final Label Shape  :", Y_all.shape)
print("Unique labels:", np.unique(Y_all, return_counts=True))
```

```python
import numpy as np

# 1. Convert labels to 0/1
Y_all = (Y_all > 0).astype(np.uint8)

# 2. Find positions
idx0 = np.where(Y_all == 0)[0]
idx1 = np.where(Y_all == 1)[0]

take1 = min(len(idx1), 60000)
take0 = take1 * 2  # 1:2 ratio

idx1s = np.random.choice(idx1, take1, replace=False)
idx0s = np.random.choice(idx0, take0, replace=False)

idx_final = np.concatenate([idx0s, idx1s])
np.random.shuffle(idx_final)

X_sample = X_all[idx_final]
Y_sample = Y_all[idx_final]

print("Sampled:", X_sample.shape, Y_sample.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_sample, Y_sample,
    test_size=0.2,
    random_state=42,
    stratify=Y_sample
)

print("Train:", X_train.shape, y_train.shape)
print("Val:", X_val.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgb

rf = RandomForestClassifier(
    n_estimators=300, max_depth=20, n_jobs=-1
)

xgb = XGBClassifier(
    n_estimators=350, max_depth=8,
    subsample=0.8, colsample_bytree=0.8,
    learning_rate=0.05,
    eval_metric="logloss"
)

lgbm = lgb.LGBMClassifier(
    n_estimators=350, num_leaves=50,
    learning_rate=0.05
)

rf.fit(X_train, y_train)
xgb.fit(X_train, y_train)
lgbm.fit(X_train, y_train)
print("RF trained")
print("XGB trained")
print("LGBM trained")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, jaccard_score

p1 = rf.predict_proba(X_val)[:,1]
p2 = xgb.predict_proba(X_val)[:,1]
p3 = lgbm.predict_proba(X_val)[:,1]

p_final = (p1 + p2 + p3) / 3
pred_bin = (p_final > 0.45).astype(int)

print("Accuracy :", accuracy_score(y_val, pred_bin))
print("Precision:", precision_score(y_val, pred_bin))
print("Recall   :", recall_score(y_val, pred_bin))
print("F1 Score :", f1_score(y_val, pred_bin))
print("IoU      :", jaccard_score(y_val, pred_bin))
```
